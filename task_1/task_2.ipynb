```python
import cv2 as cv
```

```python
cat = cv.imread('cat.jpg')
from PIL import Image
cat = cv.cvtColor(cat, cv.COLOR_BGR2RGB)
display(Image.fromarray(cat))
```

```python
left_eye = (435, 260)
right_eye = (640, 255)
ellipse_axes = (50, 35)
angle_left = 25
angle_right = 155
start_angle = 0
end_angle = 360
new_cat = cv.ellipse(cat, left_eye, ellipse_axes, angle_left, start_angle, end_angle, (255, 0, 100), 3)
cv.ellipse(new_cat, right_eye, ellipse_axes, angle_right, start_angle, end_angle, (255, 0, 100), 3)

left_ear_1 = (275, 0)
left_ear_2 = (310, 190)
left_ear_3 = (410, 85)

right_ear_1 = (760, 0)
right_ear_2 = (630, 85)
right_ear_3 = (740, 200)

nose_1 = (500, 340)
nose_2 = (600, 345)
nose_3 = (550, 400)

cv.line(new_cat, left_ear_1, left_ear_2, (255, 100, 0), 3)
cv.line(new_cat, left_ear_2, left_ear_3, (255, 100, 0), 3)
cv.line(new_cat, left_ear_1, left_ear_3, (255, 100, 0), 3)
cv.line(new_cat, right_ear_1, right_ear_2, (255, 100, 0), 3)
cv.line(new_cat, right_ear_2, right_ear_3, (255, 100, 0), 3)
cv.line(new_cat, right_ear_1, right_ear_3, (255, 100, 0), 3)
cv.line(new_cat, nose_1, nose_2, (0, 200, 0), 3)
cv.line(new_cat, nose_2, nose_3, (0, 200, 0), 3)
cv.line(new_cat, nose_1, nose_3, (0, 200, 0), 3)
display(Image.fromarray(new_cat))
new_cat = cv.cvtColor(new_cat,cv.COLOR_BGR2RGB)
cv.imwrite('new_cat.jpg', new_cat)
```
