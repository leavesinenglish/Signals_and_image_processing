```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('workbook.jpg')
from PIL import Image

img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
plt.show()
```

```python
mean_threshold = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 7, 2)
gauss_threshold = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 7, 2)
```

```python
plt.imshow(mean_threshold, cmap='gray')
cv.imwrite('mean_threshold.jpg', img)
plt.show()
```

```python
plt.imshow(gauss_threshold, cmap='gray')
cv.imwrite('gauss_threshold.jpg', img)
plt.show()
```
