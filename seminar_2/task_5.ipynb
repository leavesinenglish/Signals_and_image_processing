```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('lena.png')
from PIL import Image
img = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
cv.imwrite('lenna_wb.jpg', img)
plt.show()
```

```python
def foo(a, minv, maxv):
    if a > maxv:
        return maxv
    if a < minv:
        return minv
    return a
```

```python
N = 30

# images = np.zeros((N, img.shape[0], img.shape[1]))
def makeNoisy(img: np.array, s=300):
    image = img.copy()
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            noise = np.random.uniform(-s,s)
            val = image[i,j] + noise
            image[i,j] = foo(val, 0, 255)
    return image
```

```python
noisy = makeNoisy(img)
plt.imshow(makeNoisy(img), cmap='gray')
plt.show()
```

```python
images = []
for i in range(N):
    images.append(makeNoisy(img))
images = np.array(images)
```

```python
filtered = np.mean(images, axis=0)
plt.imshow(filtered, cmap='gray')
plt.show()
```
